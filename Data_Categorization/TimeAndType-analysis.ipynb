```python
import pandas as pd
# Carregar o arquivo CSV
df = pd.read_csv('D:/OneDrive - InMotion - Consulting/BusinessCase/Data_Categorization/processed_data.csv', encoding='utf-8')
# Exibir as primeiras linhas do dataframe para inspecionar os dados
df.head()
```

```python
import pandas as pd

# Supondo que o DataFrame já esteja carregado em 'df'

# Convertendo a coluna 'created_at' para datetime
df['created_at'] = pd.to_datetime(df['created_at'])

# Identificando a primeira e a última data
first_date = df['created_at'].min()
last_date = df['created_at'].max()

# Calculando a duração em dias, semanas, meses e anos
duration_days = (last_date - first_date).days
duration_weeks = duration_days // 7
duration_months = (last_date.year - first_date.year) * 12 + (last_date.month - first_date.month)
duration_years = last_date.year - first_date.year + 1  # +1 para incluir ambos os anos de início e fim

# Exibindo as informações
print(f"Primeira data: {first_date}")
print(f"Última data: {last_date}")
print(f"Duração em dias: {duration_days} dias")
print(f"Duração em semanas: {duration_weeks} semanas")
print(f"Duração em meses: {duration_months} meses")
print(f"Duração em anos: {duration_years} anos")

# Contagem de registros por semana e exibição do número da semana do ano
df['year_week'] = df['created_at'].dt.to_period('W').apply(lambda r: r.start_time)
df['week_number'] = df['created_at'].dt.isocalendar().week
weekly_counts = df.groupby(['year_week', 'week_number']).size()

print("\nContagem de registros por semana:")
print(weekly_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Supondo que o DataFrame já esteja carregado na variável df

# Contar o número de ocorrências de cada tipo
type_counts = df['type'].value_counts()

# Calcular o percentual de cada tipo
type_percentages = df['type'].value_counts(normalize=True) * 100

# Criar um DataFrame para visualizar os resultados
type_analysis = pd.DataFrame({
    'Count': type_counts,
    'Percentage': type_percentages
})

# Exibir a análise
print(type_analysis)
```
